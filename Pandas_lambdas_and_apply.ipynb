```python
import pandas as pd
import numpy as np
```

```python
csv_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'
df = pd.read_csv(csv_url)
df.head(
  
)
```

```python
stud_alcoh = df.loc[: , 'school': 'guardian']
```

```python
stud_alcoh.head()
```

```python
capitilizer = lambda x: x.capitalize()
```

```python
stud_alcoh['Mjob'].apply(capitilizer)
stud_alcoh['Fjob'].apply(capitilizer)
```

```python
stud_alcoh.tail()
```

```python
def majority(x):
  if x > 17:
    return True
  else:
    return False
```

```python
stud_alcoh['legal_drinker'] = stud_alcoh['age'].apply(majority)
stud_alcoh.head()
```

```python
def times10(x):
  if type(x) is int:
    return 10*x
  return x  
```

```python
stud_alcoh.applymap(times10).head(10)
```
